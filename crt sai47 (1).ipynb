```python
#include<stdio.h>
int main(){
	int a[10][10],i,j,m,n,fsum=0,bsum=0;
	scanf("%d",&n);
	for(i=0;i<n;i++){
		for(j=0;j<n;j++){
			scanf("%d",&a[i][j]);
		}
	}
for(i=0;i<n;i++){
		for(j=0;j<n;j++){
			if(i==j){
				fsum+=a[i][j];
			}
			if(i+j==(n-1)){
				bsum+=a[i][j];
			}
		}
	}
	printf("%d\n",fsum-bsum);

	
}
```
